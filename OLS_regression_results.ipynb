```python
csv_share = pd.read_csv("AwardsSharePlayers.csv")
csv_share.head()

# 'AwardsSharePlayers.csv' is recalled and assigned into 'csv_share' to print 
# out the first few lines of the dataframe. 
```

```python
merged_data_awards = csv_sal.merge(csv_share, on=["yearID", "playerID"])
merged_data_awards

# csv_sal is merged with playerID and yearID as a common key that is  
# stored in csv_share 
# to create a new dataframe called merged_data_awards 
```

```python
dfmerg_awards_share = merged_data_awards.query('yearID == 2007 and pointsWon > 30')

# the merged_data_awards is filtered with certain conditions to create 
# a new file called dfmerg_awards_share 
# 2007 was chosen in random, while 30 was a data between 50% and 75% pointsWon 
```

```python
merged_final = merged_data_awards.merge(csv_batting, on=['playerID', 'yearID'])
merged_final.head()

# merged_data_awards is merged with csv_batting on the intersection of yearID 
# and playerID 
# to create a new dataframe called merged_final 
# the purpose of the code is to merge an already merged dataframe to create 
# a new dataframe with two independent variables and one dependent variable 
```

```python
merged_final_flt = merged_final.query('yearID > 2006 and RBI > 79')
```

```python
import statsmodels.formula.api as smf
lr = smf.ols(formula='salary ~ pointsWon + RBI', data=merged_final_flt)
fitted_model = lr.fit()
fitted_model.summary()

# The R-squared value is 0.041, which is not a strong correlation because it is 
# not close to 1. 
# The number of observations is 270, which was improved by applying a 
# bigger range of yearID to show up in the dataset 
# pointsWon show that point 1 is statistically significant, so the regression 
# results show that the plots are significant 
# p>|t| values for pointsWon and RBI are below 0.05, which makes the positive  
# coefficient 95% confident 
```
